```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.io import loadmat
from scipy.ndimage import convolve,gaussian_filter
from scipy.misc import imread,imresize,imfilter
```

```python
im1_ = imread('img/baby_happy.jpg')
im2_ = imread('img/baby_weird.jpg')
im1=im1_
im2=im2_
```

```python
im1=im1_
im2=im2_
im1 = cv2.cvtColor(im1,cv2.COLOR_RGB2GRAY)
im2 = cv2.cvtColor(im2,cv2.COLOR_RGB2GRAY)
im1 = imresize(im1,(236,236))
im2 = imresize(im2,(236,236))
print(im1.shape)
print(im2.shape)
plt.subplot(1,2,1)
plt.imshow(im1)
plt.subplot(1,2,2)
plt.imshow(im2)
```

```python
im1_blur = gaussian_filter(im1,1)
im2_blur = gaussian_filter(im2,1)
plt.subplot(1,2,1)
plt.imshow(im1_blur)
plt.subplot(1,2,2)
plt.imshow(im2_blur)
```

```python
im2_detail = im2 - im2_blur
plt.subplot(1,3,1)
plt.imshow(im2,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(im2_blur,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(im2_detail,cmap='gray')
```

```python
hybrid = im1_blur+im2_detail
plt.imshow(hybrid,cmap='gray')
```

```python
im2_detail = im2.astype(np.float64) - im2_blur.astype(np.float64)
hybrid = im1_blur.astype(np.float64) + im2_detail.astype(np.float32)
plt.imshow(hybrid,cmap='gray')
```

```python
im1_blur = gaussian_filter(im1,0.5)
im2_blur = gaussian_filter(im2,0.5)
im2_detail = im2.astype(np.float64) - im2_blur.astype(np.float64)
hybrid = im1_blur.astype(np.float64) + im2_detail.astype(np.float32)
plt.imshow(hybrid,cmap='gray')
```
